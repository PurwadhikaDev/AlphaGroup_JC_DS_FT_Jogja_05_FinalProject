# A. BUSINESS UNDERSTANDING

**Produk Deposito**

Salah satu sumber pendanaan yang bisa diperoleh bank adalah dari deposito berjangka. Deposito berjangka adalah salah satu opsi investasi terbaik bagi orang yang mencari pengembalian yang stabil dan aman atas investasi mereka. Dalam deposito berjangka, sejumlah uang disimpan untuk jangka waktu tertentu dan penabung tidak diperbolehkan menarik jumlah tersebut sampai akhir periode jatuh tempo [(Sumber)](https://www.hdfcbank.com/personal/resources/learning-centre/save/term-deposit-vs-fixed-deposit).

Uang yang diterima dari nasabah akan dipinjamkan ke nasabah lain dengan bunga yang lebih tinggi. Bank akan mengambil margin dari selisih bunga tersebut karena sebagian bunga harus diberikan kepada nasabah yang membuka akun deposito.

- Bunga deposito : bunga yang diterima oleh nasabah atas uang yang mereka titipkan ke bank selama periode tertentu.
- Bunga pinjaman : bunga yang bank kenakan terhadap debitur yang meminjam uang mereka.
- Spread bunga : selisih antara bunga pinjaman dan bunga deposito yang menjadi keuntungan bagi bank. Istilah lainnya adalah *Net Interest Margin*.

Net Interest Margin (NIM) adalah selisih antara pendapatan bunga yang diperoleh oleh lembaga keuangan (misalnya bank) dari aset produktifnya dan biaya bunga yang harus dibayarkan, dinyatakan sebagai persentase dari total aset yang menghasilkan pendapatan [(Sumber)](https://www.investopedia.com/terms/n/netinterestmargin.asp).

Ada banyak cara untuk menawarkan produk perbankan termasuk doposito, yang cukup sering dilakukan adalah telemarketing. Pihak dari bank akan menawarkan produk deposito mereka kepada nasabah secara langsung melalui telepon (telemarketing).

Telemarketing dilakukan terhadap database nasabah yang dimiliki oleh bank.

**Problem Statement**

Bank HAN, merupakan salah satu bank yang ada di Portugal. Mereka memiliki data kampanye (campaign) telemarketing dari tahun 2008 - 2013 [(Sumber)](https://www.sciencedirect.com/science/article/abs/pii/S016792361400061X). Kegiatan telemarketing yang mereka lakukan untuk menawarkan deposito kepada nasabah, baik yang pernah dihubungi pada kampanye sebelumnya maupun belum pernah dihubungi. Telemarketing juga dilakukan terhadap nasabah yang sudah setuju membuka deposito pada kampanye sebelumnya dan yang belum.

Dengan cara manual, Bank HAN menemukan bahwa konversi dari hasil kegiatan telepon mereka hanya berada pada angka 11.8%, yang berarti sebagian besar kegiatan telemarketing berakhir dengan jawaban tidak dari nasabah. Bank HAN memandang ada cukup banyak waktu dan biaya yang keluar untuk menelpon nasabah yang pada akhirnya tidak membuka akun deposito. Selain itu tim marketing Bank HAN memandang bahwa biaya yang diperlukan untuk mendapatkan nilai profit tertentu dirasa terlalu besar.

Tim marketing Bank HAN merasa perlu untuk melakukan perbaikan terhadap cara mereka memilih nasabah. Dengan budget marketing yang sudah dibatasi, mereka menginginkan agar bisa mendapat daftar nasabah yang potensial untuk dihubungi dan menghindari nasabah-nasabah yang tidak akan melakukan deposito supaya biaya marketing mereka tidak terbuang sia-sia.

**Tujuan:**

- Rerata upah tahunan pekerja di Portugal periode 2008 - 2013 : 20854 euro atau setara 1737.85 per bulan [Sumber](https://www.statista.com/statistics/419498/average-annual-wages-portugal-y-on-y/)

- Biaya tenaga kerja per jam. Dengan menggunakan upah minimum 545.67 euro per bulan. Setara dengan sekitar 3.15 euro per jam, atau setara 0.0525 euro per menit.
- Asumsi jika seorang menempatkan 20% gaji tahunan untuk investasi, kemudian dengan profil risiko rendah menaruh 70% persen untuk deposito maka akan didapatkan 14% dari gaji tahunan akan masuk deposito. Setara 14% x 20854 = 2919.56 euro.

**Evaluation Metrics:**

- _True Positive (TP)_ : Nasabah yang diprediksi akan membuka deposito dan benar-benar membuka diposito.
- _False Positive (FP)_ : Nasabah yang diprediksi akan membuka deposito, ternyata tidak membuka deposito.
- _True Negative (TN)_ : Nasabah yang diprediksi tidak membuka deposito dan benar tidak membuka deposito.
- _False Negative (FN)_ : Nasabah yang diprediksi tidak membuka deposito tapi ternyata membuka deposito.


- _**Recall**_ : mengukur seberapa banyak nasabah yang membuka deposito  dan berhasil diprediksi oleh model.
  - **Rumus Recall** = TP / (TP + FN)


- _**Precision**_ : mengukur seberapa akurat model dalam memprediksi nasabah yang akan membuka deposito dibandingkan total prediksi positif yang dibuat.
  - **Rumus Precision** = TP / (TP + FP)

- **F1-Score** : mengukur seberapa baik model dalam memprediksi nasabah yang akan membuka deposito dibandingkan dengan recall dan precision.
  - **Rumus F1-Score** = 2 * (Recall * Precision) / (Recall + Precision)

**Type 1 Error** :

Type 1 error : False Positive (nasabah yang aktualnya tidak membuka deposito, tetapi diprediksi membuka deposito)

Konsekuensi : kerugian dari biaya telemarketing, ketidaknyamanan nasabah, kerugian waktu dari tenaga kerja.

**Kerugian biaya telemarketing per nasabah :**

-   Biaya sambungan telepon : Biaya telepon per menit x  median durasi (menit) = 0.076 x 3 = 0.228 Euro.

-   Biaya tenaga kerja per telp : Biaya tenaga kerja per menit 0.0525 x 3 = 0.158 Euro

-   Kerugian total biaya telemarketing per orang =  0.39 euro per nasabah

**Type 2 Error** :

Type 2 error : False Negative (nasabah yang aktualnya deposit tetapi diprediksi tidak akan deposit)
Konsekuensi: Kehilangan potensi pendapatan dari nasabah potensial

**Kerugian kehilangan potensi pendapatan dari nasabah potensial per orang:**

Asumsi deposito x margin rate : 2.919 x 1.38% = 40.28 euro per nasabah

Karena type 2 error menyebabkan kerugian yang lebih besar sehingga metriks yang diprioritaskan adalah recall untuk menekan angka false negative.

## 6 Uji Normalitas

```python
# Uji Normalitas

def check_normality(df):
    def test_normality(data):
        shapiro_p = shapiro(data).pvalue
        ks_p = kstest(data, 'norm').pvalue
        anderson_result = anderson(data)
        ad_result = anderson_result.statistic < anderson_result.critical_values[2] #using 5% critical value

        return {
            'Shapiro-Wilk': 'Normal' if shapiro_p > 0.05 else 'Not-normal',
            'Kolmogorov-Smirnov': 'Normal' if ks_p > 0.05 else 'Not-normal',
            'Anderson-Darling': 'Normal' if ad_result else 'Not-normal'
        }

    numeric_cols = df.select_dtypes(include=["number"]).columns
    results = {col: test_normality(df[col]) for col in numeric_cols}
    return pd.DataFrame.from_dict(results, orient='index').rename_axis('Feature')

normality_results = check_normality(data)
normality_results
```

Setelah melakukan uji normalitas menggunakan beberapa metode, diketahui bahwa seluruh variabel numerik tidak berdistribusi normal.

```python
# Pilih kolom numerik
numeric_columns = data.select_dtypes(include='number')
list_outliers_count = []

for column in numeric_columns.columns:
    q3 = numeric_columns.describe().loc['75%', column]
    q1 = numeric_columns.describe().loc['25%', column]
    iqr = q3 - q1
    whisker = 1.5 * iqr
    upper_bound = q3 + whisker
    lower_bound = q1 - whisker

    # Perhitungan outliers
    outliers_count = len(data[(data[column] > upper_bound) | (data[column] < lower_bound)])
    list_outliers_count.append(outliers_count)

outliers_data = pd.DataFrame({
    'Column': numeric_columns.columns,
    'Outliers Count': list_outliers_count
})
```

  - `pdays` → Jumlah hari sejak nasabah terakhir dihubungi dalam kampanye sebelumnya (999 berarti belum pernah dihubungi sebelumnya).
  - `previous` → Jumlah berapa kali nasabah dihubungi dalam kampanye sebelumnya.

**Kriteria Identifikasi Anomali**

1. Tidak masuk akal jika campaign == 0 tetapi pdays bukan 999 dan `previous` > 0.

  - campaign == 0 berarti nasabah tidak dihubungi sama sekali dalam kampanye saat ini.
  - Jika `pdays` bukan 999, berarti nasabah pernah dihubungi sebelumnya.
  - Jika `previous` > 0, berarti nasabah pernah dihubungi dalam kampanye sebelumnya.

  Kontradiksi: Bagaimana mungkin nasabah pernah dihubungi di kampanye sebelumnya (`previous` > 0) tetapi di kampanye ini tidak (`campaign` == 0)?

2. Tidak masuk akal jika `campaign` > 0, tetapi pdays == 999 dan `previous` > 0.

  - `campaign` > 0 berarti nasabah dihubungi dalam kampanye ini.
  - `pdays` == 999 berarti nasabah tidak pernah dihubungi dalam kampanye sebelumnya.
  - Jika `previous` > 0, berarti nasabah pernah dihubungi dalam kampanye sebelumnya, yang bertentangan dengan pdays == 999.

3. Tidak masuk akal jika `campaign` == 0, tetapi pdays != 999 dan `previous` == 0.

  - `campaign` == 0 berarti nasabah tidak dihubungi dalam kampanye ini.
  - `previous` == 0 berarti nasabah tidak pernah dihubungi dalam kampanye sebelumnya.
  - Tetapi jika pdays != 999, berarti nasabah pernah dihubungi sebelumnya, yang bertentangan dengan `previous` == 0.

4. Tidak masuk akal jika campaign > 0, tetapi pdays != 999 dan `previous` == 0.

  - `campaign` > 0 berarti nasabah di kampanye ini telah dihubungi beberapa kali.
  - `pdays` != 999 berarti nasabah pernah dihubungi di kampanye sebelumnya.
  - `previous` == 0 berarti nasabah tidak pernah dihubungi dalam kampanye sebelumnya, yang bertentangan dengan pdays != 999.

```python
data['anomali_cpp'] = (
    ((data['campaign'] == 0) & (data['pdays'] != 999) & (data['previous'] > 0)) |  # campaign == 0 tapi pdays bukan 999 & previous > 0
    ((data['campaign'] > 0) & (data['pdays'] == 999) & (data['previous'] > 0)) | # campaign > 0 tapi pdays == 999 & previous == 0
    ((data['campaign'] == 0) & (data['pdays'] != 999) & (data['previous'] == 0)) | # campaign == 0 tapi pdays == 999 & previous == 0
    ((data['campaign'] > 0) & (data['pdays'] != 999) & (data['previous'] == 0))  # campaign > 0 tapi pdays == 999 & previous == 0
).astype(int)
```

```python
# Menghitung nasabah
data_count = pd.crosstab(clean_data["poutcome"], clean_data["y"])

# Mengonversi ke persentase per kelompok usia
data_percent = data_count.div(data_count.sum(axis=1), axis=0) * 100

# **Balik urutan agar "85 and more" di bagian bawah**
data_count = data_count.iloc[::-1]
data_percent = data_percent.iloc[::-1]

# Membuat figure dan subplot (2 kolom)
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)  # Share Y untuk menjaga konsistensi label
colors = ["#1f77b4", "#ff7f0e"]

# ----- Subplot 1: Bar Chart Jumlah -----
ax1 = axes[0]
y_positions = np.arange(len(data_count))
bar_width = 0.4

for idx, label in enumerate(data_count.columns):
    ax1.barh(y_positions + (idx - 0.5) * bar_width, data_count[label],
             height=bar_width, label=f"{label}", color=colors[idx])

    # Menambahkan label angka
    for i, val in enumerate(data_count[label]):
        if val > 0:
            ax1.text(val + 5, y_positions[i] + (idx - 0.5) * bar_width, f"{int(val)}",
                     ha="left", va="center", fontsize=10, color="black")

ax1.set_title("Jumlah Nasabah berdasarkan Hasil Campaign Sebelumnya")
ax1.set_xlabel("Jumlah Nasabah")
ax1.set_ylabel("Hasil Campaign Sebelumnya")  # Label hanya di subplot pertama
ax1.set_yticks(y_positions)
ax1.set_yticklabels(data_count.index)
ax1.set_xlim(0, data_count.max().max() * 1.1)

# ----- Subplot 2: Stacked Bar Chart Proporsi -----
ax2 = axes[1]
left_vals = np.zeros(len(data_percent))

for idx, label in enumerate(data_percent.columns):
    ax2.barh(data_percent.index, data_percent[label], left=left_vals,
             label=f"{label}", color=colors[idx])

    # Menambahkan label angka
    for i, val in enumerate(data_percent[label]):
        if val > 0:
            ax2.text(left_vals[i] + val / 2, i, f"{val:.1f}%",
                     ha="center", va="center", fontsize=10, color="black")

    left_vals += data_percent[label].values

ax2.set_title("Proporsi Nasabah berdasarkan Hasil Campaign Sebelumnya")
ax2.set_xlabel("Proporsi Nasabah (%)")
ax2.set_xlim(0, 100)
ax2.legend(title="Membuka Deposito", labels=["Tidak", "Ya"], loc="center left", bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

**Insights :**

Variabel `poutcome` merupakan hasil campaign sebelumnya.

- Mayoritas nasabah yang menjadi target campaign adalah nasabah yang memiliki hasil campaign `nonexistent`. Hal tersebut mengindikasikan bahwa nasabah mayoritas adalah nasabah yang belum pernah dikontak oleh bank.

**Kenapa *duration* Menyebabkan Data Leakage dan Perlu Dihapus?**

Variabel duration menunjukkan lama waktu percakapan dengan nasabah selama kampanye pemasaran berlangsung. Ini menjadi sumber *data leakage* karena:

- **Diketahui Setelah Prediksi Dibuat**

  Tujuan model adalah memprediksi apakah nasabah akan berlangganan (y) sebelum mereka dihubungi.Namun, duration hanya bisa diketahui setelah panggilan selesai, sehingga informasi ini tidak tersedia pada saat prediksi seharusnya dibuat.

- **Hubungan Langsung dengan Target (y)**

  Jika durasi panggilan lama, kemungkinan besar nasabah tertarik dan akhirnya berlangganan. Sebaliknya, panggilan yang pendek bisa menunjukkan penolakan cepat.Hal ini menciptakan hubungan langsung dengan variabel target, yang tidak akan tersedia dalam skenario dunia nyata saat model digunakan.

- **Model Bisa Belajar dari Informasi Masa Depan**

  Jika duration digunakan sebagai fitur, model bisa memberikan akurasi tinggi secara tidak realistis karena memanfaatkan informasi yang seharusnya belum tersedia saat prediksi dilakukan. Ini akan menyebabkan hasil yang menyesatkan saat model diterapkan pada data baru.

## 2 Eksekusi Data Preparation

### a. Kenapa Perlu Menggunakan One-Hot Encoder, Ordinal Encoder, dan Robust Scaler?

Dalam pemrosesan data, pemilihan metode encoding dan scaling sangat penting agar model dapat memahami dan memanfaatkan data dengan benar. Berikut alasan penggunaannya:

1. **One-Hot Encoding**
  
  Digunakan untuk variabel kategorikal nominal (tanpa urutan) karena:

  Model machine learning tidak bisa langsung memahami teks.Menghindari hubungan numerik yang salah antara kategori (misalnya, "job" dengan nilai 1 dan 2 tidak berarti pekerjaan 2 lebih tinggi dari pekerjaan 1).

  **Contoh variabel: job, marital, housing, loan, contact, month, day_of_week.**

2. **Ordinal Encoding**
  
  Digunakan untuk variabel kategorikal ordinal (memiliki urutan) karena:

  Kategori memiliki hierarki yang berarti (misalnya, pendidikan dasar < menengah < tinggi). Model perlu memahami perbedaan tingkat dalam kategori.

  **Contoh variabel: age_group, education, poutcome.**

3. **Robust Scaler**

  Digunakan untuk variabel numerik dengan outlier atau skala yang bervariasi, karena:

  Tidak terpengaruh oleh outlier, berbeda dengan Standard Scaler yang bergantung pada mean dan standard deviation. Menggunakan median dan IQR (Interquartile Range) untuk normalisasi.
  
  **Contoh variabel: age, campaign, pdays, previous, emp.var.rate, cons.price.idx, cons.conf.idx, euribor3m, nr.employed.**

### a. Penjelasan Imbalanced Data

Ketidakseimbangan data dapat menyebabkan model lebih bias terhadap kelas mayoritas. Untuk mengatasi hal ini, digunakan teknik oversampling, undersampling, dan hybrid sampling.

- **Variabel Oversampling**
  
  Menambah sampel kelas minoritas agar lebih seimbang.

  1. ROS: Duplikasi acak sampel kelas minoritas.
  2. SMOTE: Membuat sampel sintetis dengan interpolasi.
  3. BorderlineSMOTE: Menambah sampel di perbatasan antar kelas.
  4. ADASYN: Fokus pada sampel kelas minoritas yang sulit diprediksi.

- **Variabel Undersampling**
  
  Mengurangi sampel kelas mayoritas untuk keseimbangan.

  1. RandomUnderSampler: Menghapus sampel kelas mayoritas secara acak.
  2. TomekLinks: Menghapus sampel mayoritas yang dekat dengan minoritas.
  3. ENN: Menghapus sampel mayoritas yang membingungkan model.
  4. NearMiss: Memilih sampel mayoritas yang paling dekat dengan minoritas.
  5. OSS: Gabungan TomekLinks dan ENN untuk seleksi sampel.

- **Variabel Hybrid Sampling**

  Menggabungkan oversampling dan undersampling.

  1. SMOTE + ENN: SMOTE untuk menambah, ENN untuk menghapus sampel mayoritas yang membingungkan.
  2. SMOTE + RUS: SMOTE untuk oversampling, RUS untuk mengurangi kelas mayoritas secara acak.

## 7 Mencari Model Terbaik Tanpa Resampling

```python
def evaluate_models(models, X_train, y_train, X_test, y_test, description, all_results):
    results = []

    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        acc = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary', pos_label=1)
        recall = recall_score(y_test, y_pred, average='binary', pos_label=1)
        f1 = f1_score(y_test, y_pred, average='binary', pos_label=1)

        results.append([name, acc, precision, recall, f1, description])

    df_results = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-score', 'Resampling Method'])

    # Tambahkan hasil evaluasi ke dalam list global
    all_results.append(df_results)

    # Ambil 4 model terbaik berdasarkan Recall
    return df_results.nlargest(4, 'Recall')

# List untuk menyimpan semua hasil evaluasi
all_results = []

# Evaluasi tanpa resampling
df_top_4 = evaluate_models(models, X_train, y_train, X_test, y_test, "Tanpa Resampling", all_results)
df_top_4
```

```python
def evaluate_models_resample(models, X_train, y_train, X_test, y_test, description, all_results):
    results = []

    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        acc = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary', pos_label=1)
        recall = recall_score(y_test, y_pred, average='binary', pos_label=1)
        f1 = f1_score(y_test, y_pred, average='binary', pos_label=1)

        results.append([name, acc, precision, recall, f1, description])

    df_results = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-score', 'Resampling Method'])

    # Ambil 1 model terbaik berdasarkan Recall
    top_1 = df_results.nlargest(1, 'Recall')

    # Tambahkan hasil evaluasi ke dalam list global
    all_results.append(top_1)

    return df_results

# List untuk menyimpan semua hasil evaluasi
all_results = []

# Evaluasi setiap metode resampling
for method, (X_resampled, y_resampled) in sampling_methods.items():
    evaluate_models_resample(models, X_resampled, y_resampled, X_test, y_test, method, all_results)

# Gabungkan semua hasil evaluasi menjadi satu DataFrame
df_all_results = pd.concat(all_results, ignore_index=True)

# Ambil 3 model terbaik berdasarkan Recall secara keseluruhan
top_3_overall = df_all_results.nlargest(3, 'Recall')

# DataFrame hasil akhir
df_all_results
```

## 9 Memproses Cross-Validation Model terbaik dari masing masing Resampling

```python
# === Cross-Validation untuk Semua Model di df_all_results ===
cv_results = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for _, row in df_all_results.iterrows():
    model_name = row['Model']

    # Ambil model yang sesuai dari dictionary models
    model = models[model_name]

    # Lakukan cross-validation dengan multiple scoring
    scores = cross_validate(model, X_train, y_train, cv=cv,
                            scoring=['accuracy', 'precision', 'recall', 'f1'])

    # Simpan hasil CV ke dalam list
    cv_results.append([
        model_name, row['Resampling Method'],
        np.mean(scores['test_accuracy']), np.std(scores['test_accuracy']),
        np.mean(scores['test_precision']), np.std(scores['test_precision']),
        np.mean(scores['test_recall']), np.std(scores['test_recall']),
        np.mean(scores['test_f1']), np.std(scores['test_f1'])
    ])

# Buat DataFrame hasil cross-validation
df_cv_results = pd.DataFrame(cv_results, columns=[
    'Model', 'Resampling Method',
    'Mean Accuracy', 'Std Accuracy',
    'Mean Precision', 'Std Precision',
    'Mean Recall', 'Std Recall',
    'Mean F1-score', 'Std F1-score'
])

# Urutkan berdasarkan Mean Recall tertinggi
df_cv_results_sorted = df_cv_results.sort_values(by='Mean Recall', ascending=False)

# Ambil 4 model terbaik berdasarkan Recall
top_4_models = df_cv_results_sorted.head(4)

# DataFrame hasil akhir (4 model terbaik)
top_4_models
```

## 15 Penjelasan Resampling & Model Terbaik

Dalam eksperimen pencarian metode resampling dan model terbaik, ditemukan bahwa kombinasi NearMiss-1 sebagai teknik under-sampling dengan XGBoost (tanpa tuning) menghasilkan model dengan nilai recall yang baik dan stabil.

Untuk memahami bagaimana kombinasi ini bekerja, berikut adalah penjelasan rinci mengenai cara kerja NearMiss-1 dan XGBoost serta bagaimana keduanya berkontribusi terhadap peningkatan performa model:

**1. NearMiss-1: Teknik Under-Sampling dengan Pemilihan Jarak Terdekat**

  NearMiss-1 adalah metode under-sampling yang secara khusus memilih sampel kelas mayoritas berdasarkan jarak terdekatnya ke sampel kelas minoritas. Teknik ini digunakan untuk mengatasi class imbalance, yaitu kondisi di mana jumlah data dalam satu kelas jauh lebih besar dibanding kelas lainnya.

  Cara Kerja NearMiss-1
  
  1. Menghitung Jarak Antar Sampel
  
    - Setiap sampel dari kelas mayoritas dihitung jaraknya ke N sampel terdekat dari kelas minoritas.

  2. Memilih Sampel dengan Jarak Rata-rata Terdekat
  
    - Hanya sampel kelas mayoritas yang memiliki rata-rata jarak terdekat ke N sampel kelas minoritas yang dipertahankan.
  
    - Sampel kelas mayoritas yang lebih jauh dari kelas minoritas akan dibuang untuk mengurangi ketidakseimbangan.

  3. Menyusun Data Baru dengan Keseimbangan Lebih Baik
  
    - Hasilnya adalah dataset yang lebih seimbang, dengan sampel kelas mayoritas yang memiliki hubungan lebih erat dengan kelas minoritas.

  Dengan metode ini, NearMiss-1 memastikan bahwa kelas mayoritas yang tersisa dalam dataset adalah sampel yang paling relevan dengan kelas minoritas, sehingga model dapat lebih fokus belajar dari pola penting yang membedakan kedua kelas.

**2. XGBoost: Algoritma Boosting untuk Meningkatkan Kinerja Model**

  XGBoost (Extreme Gradient Boosting) adalah algoritma berbasis pohon keputusan yang menggunakan pendekatan boosting untuk meningkatkan akurasi prediksi.

  Cara Kerja XGBoost

  1. Membangun Model Awal

    - Model pertama dibuat dengan pohon keputusan sederhana untuk menghasilkan prediksi awal.

  2. Menghitung Kesalahan Prediksi (Residuals)

    - Setelah model awal membuat prediksi, selisih antara prediksi dan nilai asli dihitung.
  
  3. Melatih Model Berikutnya untuk Memperbaiki Kesalahan

    - Model selanjutnya dibuat untuk memperbaiki kesalahan prediksi dari model sebelumnya.
    - Sampel yang sulit diprediksi diberi bobot lebih besar, sehingga model lebih fokus belajar dari kesalahan tersebut.

  4. Menggabungkan Model Secara Iteratif

    - Proses ini diulang beberapa kali hingga model mencapai jumlah iterasi tertentu atau perbaikan performanya tidak lagi signifikan.

**3. Peran Kombinasi NearMiss-1 dan XGBoost**

  Penggunaan NearMiss-1 sebagai teknik under-sampling memastikan bahwa distribusi kelas menjadi lebih seimbang sebelum model dilatih, sementara XGBoost mampu belajar secara optimal dari data yang telah di-resample.

  - NearMiss-1 memilih sampel kelas mayoritas yang paling relevan dengan kelas minoritas, sehingga model lebih mudah mengenali pola dari kelas minoritas.
  - XGBoost tetap mampu belajar dari dataset yang lebih kecil, karena pendekatan boosting yang berulang kali memperbaiki kesalahan prediksi.
  - Kombinasi ini menghasilkan model dengan recall yang baik dan stabil, sehingga lebih efektif dalam mendeteksi sampel dari kelas minoritas.

***Komponen Perhitungan:***
- Net Profit = Total Pendapatan Bank - Total Biaya Telepon
- Total Pendapatan Bank = Net Interest Margin Total Deposito Diperoleh
- Total biaya terdiri dari biaya jaringan telepon dan biaya tenaga kerja.
- Biaya jaringan = Median durasi telepon (menit) Biaya telepon per menit
- Biaya tenaga kerja = Median durasi telepon (menit) Biaya tenaga kerja per menit
- Total biaya telepon per nasabah = Biaya jaringan + Biaya tenaga kerja
- Total biaya telepon = Total biaya telepon per nasabah Jumlah nasabah yang membuka deposito

**Daftar Komponen Biaya, Rerata Deposito per Nasabah & Net Interest Margin (NIM):**
- Biaya telepon per menit (euro) = 0.076
- Median durasi telepon (menit) = 3
- Biaya jaringan per nasabah = euro
- Biaya tenaga kerja per menit (euro) = 0.0525
- Biaya tenaga kerja per nasabah = euro
- Total biaya telepon per nasabah = euro
- Net Interest Margin (NIM) = 1.38%
- Rerata deposito per nasabah (euro) = 2919

**Tanpa Menggunakan Machine Learning**

- Jumlah nasabah yang ditelpon = 9628 (semua nasabah ditelpon)
- Jumlah nasabah membuka deposito = 1133
- Total Deposito Diperoleh = 1133 x 2919 = 3,307,227 euro
- Total Pendapatan Bank = 1.38% x 3,307,227 = 45,639.73 euro
- Total Biaya Telepon = 9628 x 0.3855 = 3,711.59 euro
- Net Profit = 45,639.73 - 3,711.59 = 41,928.14 euro
- Cost to Profit Ratio = 8.85%

**Menggunakan Machine Learning**
Confusion Matrix
- True Positive (TP) = 919
- False Positive (FP) = 5312
- True Negative (TN) = 3183
- False Negative (FN) = 214

**Perhitungan**
- Jumlah nasabah yang ditelpon = TP + FP = 919 + 5312 = 6231
- Jumlah nasabah membuka deposito = TP = 919
- Total Deposito Diperoleh = 919 x 2919 = 2,681,841 euro
- Total Pendapatan Bank = 1.38% x 2,681,841 = 37,009.41 euro
- Total Biaya Telepon = 6231 x 0.3855 = 2,402.76 euro
- Net Profit = 37,009.41 - 2,402.76 = 34,606.65 euro
- Cost to Profit Ratio = 6.95%

# Rekomendasi untuk Bisnis

## 1. Waktu dan Frekuensi Kontak
- Fokuskan promosi pada bulan dengan konversi tinggi: **Maret, April, September, Oktober, Desember**.
- Prioritaskan hari **Kamis** untuk telemarketing (konversi tertinggi).
- Hindari hari **Senin** dan **Jumat** (konversi rendah).
- Lakukan follow-up dalam **3-6 hari** setelah kontak pertama.
- Batasi jeda kontak >7 hari (konversi menurun drastis).
- Lakukan **1-3 kali percobaan kontak** untuk meningkatkan peluang konversi.
- Gunakan **komunikasi via cellular** (konversi lebih tinggi).
- Hindari menghubungi nasabah yang tidak subscribe sebelumnya, batasi kontak maksimal **3 kali** untuk menghindari spam.

## 2. Segmentasi Nasabah dan Strategi Pendekatan
### Kelompok Usia
- Fokus pada **usia muda (15-24 tahun)** dan **usia tua (>60 tahun)**.
  - **Usia muda**: Tawarkan produk relevan dan edukatif.
  - **Usia tua**: Tekankan keamanan dan stabilitas dana.

### Pendidikan
- **Pendidikan tinggi (`university.degree`)**: Optimalkan pendekatan edukasi.
- **Pendidikan menengah (`high.school`, `professional.course`)**: Berikan informasi manfaat deposito.
- **Pendidikan rendah**: Gunakan pendekatan sederhana dan mudah dipahami.

### Pekerjaan
- **Pelajar (`student`)**: Tawarkan produk menarik.
- **Pensiunan (`retired`)**: Tekankan keamanan dan keuntungan jangka panjang.

### Status Pernikahan
- **Lajang (`single`)**: Fokus pada keuntungan jangka panjang dan fleksibilitas investasi.

## 3. Strategi Telemarketing Efektif
- Durasi panggilan **optimal**: Cukup lama untuk menarik dan informatif, tetapi tidak terlalu lama (>300 detik menurunkan konversi).
- Jika nasabah setuju cepat, kemungkinan sudah teredukasi atau skrip telemarketing efektif.
- Jika panggilan berakhir terlalu cepat tanpa hasil positif, nasabah kemungkinan sudah memutuskan menolak.

## 4. Faktor Ekonomi yang Mempengaruhi Deposito
- **Euribor3m tinggi**: Tawarkan suku bunga deposito lebih tinggi.
- **CCI tinggi**: Tekankan keamanan dan pertumbuhan dana jangka panjang.
- **CPI tinggi (inflasi)**: Tekankan stabilitas keuangan deposito.
- **Employment Variation Rate positif**: Fokus pada segmen baru bekerja atau berpenghasilan tetap.
- **Jumlah tenaga kerja tinggi**: Perkenalkan deposito sebagai cara efektif menyimpan dan mengembangkan dana.
